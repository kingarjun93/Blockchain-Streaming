```
from websocket import create_connection
ws = create_connection("wss://ws.blockchain.info/inv")
print("Sending 'Hello, World'...")
ws.send("{\"op\":\"unconfirmed_sub\"}")
print( "Sent")
print( "Receiving...")
i=0
fullresult = []
while(i<100):
  result =  ws.recv()
  #print( "Received '%s'" % result)
  fullresult.append(result)
  print(i, ' page completed')
  i=i+1
ws.close()
```

```
%sql

--SELECT distinct to_timestamp(x.time) FROM test
WITH inputs AS (
SELECT to_timestamp(x.time) AS time_of_transaction,x.hash AS hash, explode(x.inputs) AS in FROM test)
SELECT in.prev_out.addr AS Address,
in.prev_out.value AS SatoshiAmount,
time_of_transaction,
hash
FROM inputs
ORDER BY time_of_transaction DESC
LIMIT 100;
```

```
%sql

--SELECT distinct to_timestamp(x.time) FROM test
WITH inputs AS (
SELECT to_timestamp(x.time) AS time_of_transaction,x.hash AS hash, explode(x.inputs) AS in FROM test),
trans AS(SELECT in.prev_out.addr AS Address,
in.prev_out.value AS SatoshiAmount,
time_of_transaction,
minute(time_of_transaction)||':'||second(time_of_transaction) AS minute,
hash
FROM inputs
WHERE time_of_transaction >= (current_timestamp - INTERVAL 1 HOUR)
)
SELECT minute,COUNT(*) AS counts
FROM trans
GROUP BY minute
HAVING counts > 10 --replace this value with min value
ORDER BY minute
```

```
%sql

WITH inputs AS (
SELECT to_timestamp(x.time) AS time_of_transaction,x.hash AS hash, explode(x.inputs) AS in FROM test),
trans AS
(SELECT in.prev_out.addr AS Address,
in.prev_out.value AS SatoshiAmount,
time_of_transaction,
hash
FROM inputs
WHERE time_of_transaction >= (current_timestamp - INTERVAL 3 HOUR)
)
SELECT address,SUM(SatoshiAmount) AS total_value 
FROM trans
GROUP BY address
ORDER BY 2 DESC
```
